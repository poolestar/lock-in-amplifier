```python
import matplotlib.pyplot as plt
import numpy as np
import random
import scipy.fftpack
```

```python
sampling_rate = 1
number_samples = 1000
```

```python
frequency = 0.1
amplitude = 1
t = np.linspace(0.0, number_samples, number_samples)
signal = amplitude*np.sin(2 * np.pi * frequency * t / sampling_rate)
```

```python
plt.figure(figsize=(16, 4))
#plt.subplot(1, 2, 1)
plt.plot(t, signal)
plt.ylabel('Amplitude')
plt.xlabel('Samples')
plt.show()
```

```python
x = np.linspace(0.0, sampling_rate/2, number_samples)
signal_fft = scipy.fftpack.rfft(signal)
#signal_fft = scipy.fftpack.fftshift(signal_fft)
signal_fft = 2.0*np.abs(signal_fft[:number_samples])/number_samples
#signal_fft = 10*np.log10(signal_fft)
plt.figure(figsize=(16, 4))
plt.plot(x, signal_fft)
plt.show()
```

```python
N = number_samples
x = signal
fs = sampling_rate
sp = np.fft.rfft(x)
freq = np.arange((N / 2) + 1) / (float(N) / fs)
s_mag = np.abs(sp) * 2 / N
s_dbfs = 10 * np.log10(s_mag/1)
plt.figure(figsize=(16, 4))
plt.plot(freq,s_mag)
plt.show()
```

```python
# Seed the random number generator
np.random.seed(1234)

Fs = 1
frequency = 0.1
N = 100

time_vec = np.arange(0, N, 1)
print(len(time_vec))
signal = np.sin(2 * np.pi * frequency * time_vec/Fs) + 0.00 * np.random.randn(time_vec.size)

plt.figure(figsize=(16, 4))
plt.plot(time_vec, signal)
plt.show()
```

```python
# The FFT of the signal
sig_fft = scipy.fftpack.fft(signal)/N
# And the power (sig_fft is of complex dtype)
magnitude = np.abs(sig_fft)
power_db = 20*np.log10(magnitude)
# The corresponding frequencies
sample_freq = scipy.fftpack.fftfreq(signal.size, 1)
sample_freq1 = np.linspace(-0.5, 0.5, N, False)
print(sample_freq)
print(sample_freq1)
#print(magnitude)
#print(len(sample_freq))
#print(len(magnitude))
# Plot the FFT Magnitude
plt.figure(figsize=(16, 4))
plt.plot(sample_freq1, power_db)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.show()
```
