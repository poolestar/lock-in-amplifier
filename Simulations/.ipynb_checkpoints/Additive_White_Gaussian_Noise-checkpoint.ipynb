```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import fft, fftshift, rfft
import random
```

```python
fs = 1
N = 1000
t = np.linspace(0, N/fs, N, endpoint=False)
print(t[0:10:1])
```

```python
f = 0.1
a = 1
dc = 0
signal = a*np.sin(2*np.pi*f*t/fs) + dc
```

```python
plt.figure(figsize=(16, 4))
#plt.subplot(1, 2, 1)
plt.plot(signal)
plt.ylabel('Amplitude')
plt.xlabel('Samples')
plt.show()
```

```python
Es = 1/(len(signal))*np.sum(np.abs(np.square(signal)))
print(Es)
```

```python
SNR_dB = 30
SNR_lin = np.power(10,(SNR_dB/10))
print(SNR_lin)
```

```python
noise = np.sqrt(Es/1)*np.random.randn(1,len(signal))
#print(noise)
```

```python
plt.figure(figsize=(16, 4))
#plt.subplot(1, 2, 1)
plt.plot(noise)
plt.ylabel('Amplitude')
plt.xlabel('Samples')
plt.show()
```
