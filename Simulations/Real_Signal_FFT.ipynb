```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftshift, rfft
```

```python
fs = 1
N = 1000
t = np.linspace(0, N/fs, N, endpoint=False)
print(t[0:10:1])
```

```python
plt.figure(figsize=(16,2))
#plt.subplot(1, 2, 1)
plt.plot(t, linestyle = 'None', marker='.')
plt.show()
```

```python
f = 0.1
a = 1
dc = 1
signal = a*np.sin(2*np.pi*f*t/fs) + dc
```

```python
plt.figure(figsize=(16,2))
plt.plot(signal)
plt.ylabel('Amplitude')
plt.xlabel('Samples')
plt.show()
```

```python
x_fft = np.linspace(0, fs/(2.0), N, endpoint=False)
print(len(x_fft))
#print(x_fft)
```

```python
plt.figure(figsize=(16,2))
plt.plot(x_fft, linestyle = 'None', marker='.')
plt.grid()
plt.show()
```

```python
signal_fft = rfft(signal)
signal_fft_magnitude = np.append(np.abs(signal_fft[0])/N,np.abs(signal_fft[1:])*2/N)
signal_fft_power = np.square(signal_fft_magnitude)
signal_fft_power_db = 10*np.log10(signal_fft_power)
#signal_fft = fftshift(signal_fft)
print(len(signal_fft))
#print(signal_fft_magnitude)
```

```python
plt.figure(figsize=(16,2))
plt.plot(x_fft, signal_fft_power_db)
plt.grid()
plt.show()
```
